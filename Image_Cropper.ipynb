```python
# Image Cropper to crop off any face that it detects

# import image utilities
from PIL import Image
import os
import cv2
import numpy as np
import dlib
from tqdm import tqdm

# define a function that rotates images in the current directory
# given the rotation in degrees as a parameter
def rotateImages(rotationAmt, path):
  # for each image in the current directory
  #for image in os.listdir("D:/dataser/28/10"):
    # open the image
    img = Image.open(path)
    # rotate and save the image with the same filename
    img.rotate(rotationAmt).save(path)
    # close the image
    img.close()
    
def is_empty(faces):
    x = 0
    for face in faces:
        x += 1
    if x == 0:
        return True
    return False
    
def faceDetection(path):
    detector = dlib.get_frontal_face_detector()
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)
    return faces

def imgCrop(path):
    faces = faceDetection(path)
    if not is_empty(faces):
        #delete image that face cannot be detected
        #os.remove(path)
        img = Image.open(path)
        for face in faces:
            x1 = face.left() # left point
            y1 = face.top() # top point
            x2 = face.right() # right point
            y2 = face.bottom() # bottom point
        
        temp = img.crop((x1,y1,x2,y2))
        temp.save(path)
        img.close()
    else:
        print("Face cannot detected at " + path)
    

            
def test(path):
    imgCrop(path)

path = "D:/first 1 mintue/"

#folder = "5"
for i in range(50,61):
    print("Currently working on person " + str(i) + "!!!!")
    for folder in os.listdir(path + str(i)):
        print("Currently in " + folder)
        for image in os.listdir(path + str(i) + "/" + folder):
            imgCrop(path + str(i) + "/" + folder + "/" + image)
        print("Finished" + folder)
    print("Done with person" + str(i) + " !")
```

```python
def non_cropped_filter(path):
    for i in tqdm(os.listdir(path)):
        print("Currently working on person " + str(i) + "!!!!")
        for folder in os.listdir(path + str(i)):
            print("Currently in " + folder)
            for image in os.listdir(path + str(i) + "/" + folder):
                im = Image.open(path + str(i) + "/" + folder + "/" + image)
                width, height = im.size
                if width > 800 or height > 800:
                    im.close()
                    os.remove(path + str(i) + "/" + folder + "/" + image)
                    print("***** removed *****" + "/" + folder + "/" + image)
                else:
                    im.close()
            print("Finished" + folder)
    print("Done with person" + str(i) + " !")
    
path = "D:/first 1 mintue/"

non_cropped_filter(path)
```
